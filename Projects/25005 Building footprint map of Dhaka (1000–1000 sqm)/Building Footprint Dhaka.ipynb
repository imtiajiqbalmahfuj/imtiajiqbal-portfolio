# osmnx is the openstreetmap python library
# contextily is used to add basemap to the osm data fetched

# Importing the necessary libraries
import osmnx as ox
import contextily as ctx

# Area of interest 
address = 'Kumasi, Ghana'
# interested in the building data in osm
tags = {'building':True}

# Geocoding the place of interest
# Getting building data within  1000m by 1000m
buildings = ox.features_from_address(address, 
 tags = tags, 
 dist= 1000 )

# reprojecting to web mercator in order to add the basemap
osm_basemap = buildings.to_crs('epsg:3857')

# plotting the map
ax = osm_basemap.plot(figsize=(10,10),
 column='building',
 cmap='tab20',
 legend=True,
 legend_kwds={'loc':'upper left','bbox_to_anchor':(1,1),'frameon':False})
# other futures
ax.set_title('Building Types in ' + address, fontweight = 'bold')
ax.axis('off')

# Adding a basemap
ctx.add_basemap(ax, source=ctx.providers.CartoDB.DarkMatter)

```python
# Importing the necessary libraries
import osmnx as ox
import contextily as ctx
import matplotlib.pyplot as plt
```

```python
# Area of interest 
address = 'Purbachal, Bangladesh'
# interested in the building data in osm
tags = {'building':True}
```

```python
# Geocoding the place of interest
# Getting building data within  1000m by 1000m
buildings = ox.features_from_address(address, 
 tags = tags, 
 dist= 4000 )
```

```python
# reprojecting to web mercator in order to add the basemap
osm_basemap = buildings.to_crs('epsg:3857')
```

```python
# plotting the map
ax = osm_basemap.plot(figsize=(10,10),
 column='building',
 cmap='tab20',
 legend=True,
 legend_kwds={'loc':'upper left','bbox_to_anchor':(1,1),'frameon':False})

# other futures
ax.set_title('Building Types in ' + address, fontweight = 'bold')
ax.axis('off')

# Adding a basemap
ctx.add_basemap(ax, source=ctx.providers.CartoDB.DarkMatter)

# Save the figure with 300 PPI
plt.savefig("Dhaka_1000m_building_map.jpg", dpi=300, format='jpg', bbox_inches='tight')
```
