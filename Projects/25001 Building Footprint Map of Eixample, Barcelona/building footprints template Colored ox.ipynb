```python
import osmnx as ox
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
cmap = mcolors.LinearSegmentedColormap.from_list('blue_pink', ['#D4145A', '#FBB03B'])
cmap
```

```python
admin_district = ox.geocode_to_gdf('Eixample, Barcelona')
admin_poly = admin_district.geometry.to_list()[0]
admin_poly
```

```python
footprints = ox.features_from_polygon(admin_poly, tags = {'building' : True})
```

```python
len(footprints)
```

```python
f, ax = plt.subplots (1,1, figsize = (10,10))
footprints.plot(ax=ax, cmap = cmap, linewidth = 1.0)
ax.axis ('off')
f.patch.set_facecolor ('black')
#plt.savefig('building footprint of Eixample Barcelona 2.png', dpi =500)
#plt.savefig('building footprint of Eixample Barcelona 2.pdf', dpi =500)
```
