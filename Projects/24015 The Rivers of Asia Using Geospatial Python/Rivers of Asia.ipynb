### Plotting

```python
f, ax = plt.subplots (1,1, figsize = (16,12))

rivers ['width'] = 10 - rivers ['A_Strahler']
rivers ['width'] = rivers ['width'] / 7.0

rivers.plot(
    ax = ax, 
    column = 'A_Strahler', 
    linewidth = rivers ['width'],
    cmap = 'Blues'
)

f.patch.set_facecolor ('k')
ax.set_facecolor('k')
ax.axis('off')
ax.set_title(
    "Rivers of Asia",
    fontsize = 18,
    color = 'Blue',
    weight = 'bold',
    fontfamily = 'serif',
    y=.98,
)
plt.text(
    .95, 0.02, "Created By: Imtiaj Iqbal Mahfuj",
    fontsize=10, color='white',
    transform=ax.transAxes, 
    ha='right', va='bottom', family='monospace'
)
plt.savefig("Rivers_of_asia.pdf", dpi = 500)
plt.savefig("Rivers_of_asia.png", dpi = 500)
```
