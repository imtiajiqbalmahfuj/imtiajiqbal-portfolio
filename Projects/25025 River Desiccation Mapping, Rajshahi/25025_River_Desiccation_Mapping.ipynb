## üìù Overview
This project investigates the **desiccation of the Padma River** over a 25-year period (2001‚Äì2025), focusing on the **Rajshahi region** of Bangladesh. The Padma River, one of the most critical river systems in South Asia, has shown significant **hydrological and morphological changes** in recent decades due to both natural and anthropogenic pressures.  

By leveraging **MODIS Surface Reflectance data (MOD09A1)**, this study computes the **Normalized Difference Water Index (NDWI)** annually to assess temporal changes in surface water extent.  

The main objectives include:
- Quantifying long-term changes in the Padma River‚Äôs surface water.
- Visualizing annual variations in NDWI to identify desiccation trends.
- Demonstrating how open-source geospatial Python tools can be used for time-series river analysis.  

This analysis is relevant for **hydrologists, climate researchers, planners, and policy-makers** seeking to understand **river health**, **climate impacts**, and **sustainable water management** strategies.

## ‚öôÔ∏è Methodology
| Step | Description |
|------|-------------|
| **1. Data Collection** | MODIS Surface Reflectance (MOD09A1) from Google Earth Engine (2001‚Äì2025). |
| **2. Preprocessing** | Selected NIR (`sur_refl_b02`) and Green (`sur_refl_b04`) bands, clipped to Rajshahi AOI. |
| **3. Analysis** | Computed NDWI = (Green - NIR) / (Green + NIR) for each timestep; aggregated annually. |
| **4. Visualization** | Used `xarray` and `matplotlib` to plot 25 annual NDWI frames showing water area changes. |

## Codes

```python
import ee
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import xee
```

```python
ee.Authenticate()
ee.Initialize()
```

```python
modis = (
    ee.ImageCollection("MODIS/061/MOD09A1")\
    .select('sur_refl_b02','sur_refl_b04')\
    .filterDate('2001','2025')
)
```

```python
aoi = ee.Geometry.Rectangle([88.45694402418187, 24.057944340289367, 
                             89.05157903394749, 24.392323604959078])
```

```python
ds = xr.open_dataset(modis, engine = 'ee', crs = 'EPSG:4326', scale= 0.01, geometry= aoi)
```

```python
ds
```

```python
green = ds.sur_refl_b04 * 0.0001
nir = ds.sur_refl_b02 * 0.0001
```

```python
nir
```

```python
ndwi = (green - nir) / (green + nir)
```

```python
ndwi
```

```python
ndwi_annual = ndwi.resample(time = 'Y').mean('time')
```

```python
ndwi_annual
```

```python
ndwi_annual.plot(x='lon', y= 'lat', cmap= 'Blues', col = 'time', col_wrap = 5, robust = True)
plt.suptitle("River Desiccation Mapping of Rajshahi using Python", x=.45, y=1.03, size=24, weight='bold')
plt.figtext(0.8, 0.045, 'Imtiaj Iqbal', 
            horizontalalignment='right', 
            fontsize=14, 
            color='gray')
plt.savefig('River_Desiccation.png', dpi = 360, bbox_inches='tight')
```
